```python
import os
import pandas as pd
import numpy as np
import cv2
import pytesseract
import re
from tqdm import tqdm
from concurrent.futures import ThreadPoolExecutor, TimeoutError
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, LSTM, Embedding
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Set environment variable to avoid fork() warning
os.environ['OPENBLAS_NUM_THREADS'] = '1'

sys.path.append('/content/res/src')
from utils import download_images
from constants import entity_unit_map

# Function to safely download images
def safe_download_image(url, timeout=30):
    try:
        return process_with_timeout(download_images, url, timeout=timeout)
    except Exception as e:
        print(f"Failed to download image from {url}: {str(e)}")
        return None

# Function to process with timeout
def process_with_timeout(func, *args, timeout=60):
    with ThreadPoolExecutor(max_workers=1) as executor:
        future = executor.submit(func, *args)
        try:
            return future.result(timeout=timeout)
        except TimeoutError:
            print(f"Function {func.__name__} timed out after {timeout} seconds")
            return None

# Image preprocessing function
def preprocess_image(image_path, target_size=(224, 224)):
    try:
        img = cv2.imread(image_path)
        if img is None:
            raise ValueError(f"Failed to read image: {image_path}")
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img = cv2.resize(img, target_size)
        img = img / 255.0  # Normalize
        return img
    except Exception as e:
        print(f"Error preprocessing image {image_path}: {str(e)}")
        return None

# OCR function
def extract_text(image_path):
    try:
        img = cv2.imread(image_path)
        if img is None:
            raise ValueError(f"Failed to read image: {image_path}")
        text = pytesseract.image_to_string(img)
        return text
    except Exception as e:
        print(f"Error extracting text from {image_path}: {str(e)}")
        return ""

# Entity extraction function
def extract_entity(text, entity_name):
    patterns = {
        'item_weight': r'(\d+(?:\.\d+)?)\s*(gram|kilogram|ounce|pound)',
        # Add patterns for other entity types
    }

    if entity_name in patterns:
        matches = re.findall(patterns[entity_name], text, re.IGNORECASE)
        if matches:
            value, unit = matches[0]
            if unit.lower() in entity_unit_map[entity_name]:
                return f"{float(value)} {unit.lower()}"
    return ""

# Load data
train_df = pd.read_csv('/content/res/dataset/train.csv')
test_df = pd.read_csv('/content/res/dataset/test.csv')

print(train_df.head())
print(test_df.head())

# Process training data
processed_data = []

for _, row in tqdm(train_df.iterrows(), total=len(train_df), desc="Processing training data"):
    try:
        image_path = safe_download_image(row['image_link'])
        if image_path is not None:
            preprocessed_image = preprocess_image(image_path)
            if preprocessed_image is not None:
                text = extract_text(image_path)
                entity = extract_entity(text, row['entity_name'])
                processed_data.append({
                    'image': preprocessed_image,
                    'text': text,
                    'entity_name': row['entity_name'],
                    'entity_value': row['entity_value']
                })
    except Exception as e:
        print(f"Error processing row {row['index']}: {str(e)}")
        continue

# Prepare data for model
X_text = [d['text'] for d in processed_data]
y = [d['entity_value'] for d in processed_data]

tokenizer = Tokenizer()
tokenizer.fit_on_texts(X_text)
X_seq = tokenizer.texts_to_sequences(X_text)
X_pad = pad_sequences(X_seq, maxlen=100)

# Define model
input_text = Input(shape=(100,))
x = Embedding(len(tokenizer.word_index) + 1, 64)(input_text)
x = LSTM(64)(x)
output = Dense(1)(x)

model = Model(inputs=input_text, outputs=output)
model.compile(optimizer='adam', loss='mse')

# Train model
model.fit(X_pad, y, epochs=10, validation_split=0.2, verbose=1)

# Process test data
test_processed = []

for _, row in tqdm(test_df.iterrows(), total=len(test_df), desc="Processing test data"):
    try:
        image_path = safe_download_image(row['image_link'])
        if image_path is not None:
            text = extract_text(image_path)
            test_processed.append({
                'index': row['index'],
                'text': text,
                'entity_name': row['entity_name']
            })
    except Exception as e:
        print(f"Error processing test row {row['index']}: {str(e)}")
        continue

X_test_text = [d['text'] for d in test_processed]
X_test_seq = tokenizer.texts_to_sequences(X_test_text)
X_test_pad = pad_sequences(X_test_seq, maxlen=100)

# Make predictions
predictions = model.predict(X_test_pad)

# Format predictions
def format_prediction(pred, entity_name):
    # Implement logic to format prediction according to entity type
    # Use entity_unit_map to ensure correct units
    # This is a placeholder implementation
    return f"{pred:.2f} {list(entity_unit_map[entity_name])[0]}"

output_data = []
for pred, data in zip(predictions, test_processed):
    formatted_pred = format_prediction(pred[0], data['entity_name'])
    output_data.append({
        'index': data['index'],
        'prediction': formatted_pred
    })

output_df = pd.DataFrame(output_data)
output_df.to_csv('test_out.csv', index=False)

# Run sanity check
!python src/sanity.py test_out.csv
```
